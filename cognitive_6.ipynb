```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Mx²', linestyle='--', color='b')
plt.plot(x, y2, label='Msin(x)', linestyle='-', color='r')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Plot of y = Mx² and y = Msin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({'Subjects': ['Math', 'Science', 'English', 'History', 'Computer'],
                   'Scores': [85, 90, 78, 88, 95]})

ax = sns.barplot(x='Subjects', y='Scores', hue='Subjects', data=df, palette='viridis', legend=False)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.title('Scores by Subject')
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(102317266)

data = np.random.randn(50)
cumsum_data = np.cumsum(data)

fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(cumsum_data, marker='o')
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), data, color='red')
axes[0, 1].set_title('Random Scatter')
axes[0, 1].grid(True)

axes[1, 0].axis('off')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
