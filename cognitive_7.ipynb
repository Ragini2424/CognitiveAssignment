```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

roll_number = 102317266
np.random.seed(roll_number)

sales_data = np.random.randint(1000, 5001, size=(12, 4))

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
df = pd.DataFrame(sales_data, columns=categories, index=months)

print(df.head())
print(df.describe())

total_sales_per_category = df.sum()
total_sales_per_month = df.sum(axis=1)

growth_rates = df.pct_change().mean()

df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

if roll_number % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.90
else:
    df["Clothing"] = df["Clothing"] * 0.85

plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(df.index, df[category], label=category)
plt.legend()
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_values = np.abs(array)
print("Element-wise Absolute Values:\n", abs_values)

percentile_flat = np.percentile(array, [25, 50, 75], axis=None)
percentile_col = np.percentile(array, [25, 50, 75], axis=0)
percentile_row = np.percentile(array, [25, 50, 75], axis=1)

print("\nPercentiles (Flattened):", percentile_flat)
print("Percentiles (Per Column):\n", percentile_col)
print("Percentiles (Per Row):\n", percentile_row)

mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nMean (Flattened):", mean_flat, "Median:", median_flat, "Std Dev:", std_flat)
print("Mean (Per Column):", mean_col)
print("Median (Per Column):", median_col)
print("Std Dev (Per Column):", std_col)
print("Mean (Per Row):", mean_row)
print("Median (Per Row):", median_row)
print("Std Dev (Per Row):", std_row)
```

```python
import numpy as np
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Original Array:", a)
floor_values = np.floor(a)
print("Floor Values:", floor_values)
ceiling_values = np.ceil(a)
print("Ceiling Values:", ceiling_values)
truncated_values = np.trunc(a)
print("Truncated Values:", truncated_values)
rounded_values = np.round(a)
print("Rounded Values:", rounded_values)
```
